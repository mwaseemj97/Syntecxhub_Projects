```python
df_sorted = df['Sales'].sort_values(ascending = False)
df.groupby(df['Region'])
```

```python
plt.figure(figsize = (10,4))
plt.plot(df['Date'], df['Sales'])
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Date vs Sales')
plt.xticks(rotation = 'vertical', size = 8)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.figure(figsize = (5,5))
plt.bar(df['Category'], df['Sales'])
plt.title('Category Vs. Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()


plt.subplot(1,2,2)
plt.pie(df['Sales'],labels=df['Region'])
plt.title("Sales Distribution")
plt.show()

```

```python
plt.subplot(1,2,1)
plt.figure(figsize = (5,5))
plt.bar(df['Category'], df['Sales'])
plt.title('Category Vs. Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.savefig('1')
plt.show()


plt.subplot(1,2,2)
plt.pie(df['Sales'],labels=df['Region'])
plt.title("Sales Distribution")
plt.savefig('2')
plt.show()
```
