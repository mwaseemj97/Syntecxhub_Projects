# Exploratory Data Analysis

<h4> What is the Overall Sales Trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year from the dataset

df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
# Grouping month year

df_trend = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
df_trend
```

```python
# Set the figure size 
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation='vertical', size = 8)
plt.savefig('month_trend')
plt.show()
```

<h2> What are the top 10 Products by Sales?

```python
# Grouping the product name column
prod_sales =pd.DataFrame(df.groupby('product_name')['sales'].sum())
```

```python
prod_sales
```

```python
prod_sales = prod_sales.sort_values('sales', ascending = False)
```

```python
prod_sales[:10]
```

<h4> Which are the most selling products?

```python
# Grouping product names

most_sell_prod = pd.DataFrame(df.groupby('product_name')['quantity'].sum())
```

```python
# Sorting most_sell_products

most_sell_prod = most_sell_prod.sort_values('quantity', ascending=False)
```

```python
most_sell_prod
```

```python
most_sell_prod[:10]
```

- <h4> What is the most prefered shipping method?

```python
# Setting figure size 

plt.figure(figsize=(10,8.5))

# Plotting Shipment
sns.countplot(x='ship_mode', data=df, palette='colorblind')
plt.title('Orders by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Number of Orders')
plt.savefig('Orders by Ship Mode')
plt.show()
```

- <h4> Which are the most profitable category and sub-category?

```python
# Grouping Category and Subcategory

cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category'])['profit'].sum())
```

```python
# Sorting the results

cat_subcat_profit.sort_values(['category', 'profit'], ascending = False)
```
